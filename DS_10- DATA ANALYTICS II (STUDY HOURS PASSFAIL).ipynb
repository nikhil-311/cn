```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

# Dataset given is different from the question asked in the problem statement

# Hence refer to practical no. 9

```python
df = pd.DataFrame({
    'STUDY HOURS':[29, 15, 33, 48, 39],
    'PASS/FAIL':[0,0,1,1,1] 
})
```

```python
X = df[['STUDY HOURS']]
y = df['PASS/FAIL']
```

```python
model = LogisticRegression()
```

```python
model.fit(X,y)
```

```python
def result(n):
    n = pd.DataFrame([[n]], columns=['STUDY HOURS'])
    output = model.predict(n)[0]
    res = 'pass' if output==1 else 'fail'
    return res
```

```python
# taking inputs
result(int(input("Enter the study hours:")))
```
