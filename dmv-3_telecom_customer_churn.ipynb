```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\Acer\Desktop\dmv and ml\DMV_praticals\Datasets\telecom_customer_churn (1).csv")
data.head()

data.describe()
data.info()

data.isnull().sum()
data.dropna(inplace= True)
data.isnull().sum()
data

data.columns
# List of service-related columns
services = [
    'Phone Service',
    'Multiple Lines',
    'Internet Service',
    'Online Security',
    'Online Backup',
    'Device Protection Plan',
    'Premium Tech Support',
    'Streaming TV',
    'Streaming Movies'
]


data['Total Services'] = data[services].apply(lambda x: x.str.contains('Yes').sum(), axis=1)

data['High Data User'] = data['Avg Monthly GB Download'].apply(lambda x: 1 if x> data['Avg Monthly GB Download'].mean() else 0)

data['Referral Status'] = data['Number of Referrals'].apply(lambda x: 1 if x > 0 else 0)

data.columns
data

from urllib.request import DataHandler
from sklearn.preprocessing import MinMaxScaler,StandardScaler

scaler = MinMaxScaler()
numerical_columns = data.select_dtypes(include=['number']).columns
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

data.head()

data.isnull().sum() 

```
